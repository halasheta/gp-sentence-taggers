```python
import nltk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
nltk.download('brown')
nltk.download('averaged_perceptron_tagger')
from nltk.corpus import brown
```

```python
txt = brown.tagged_words()
```

```python
tokens = [(tup[0].lower(), tup[1]) for tup in txt]
word_freq = {}

# Make a word-frequency dict
for token in tokens:
    if token[0] not in word_freq:
        word_freq[token[0]] = {}
        
    if token[1] not in word_freq[token[0]]:
        word_freq[token[0]][token[1]] = 1
    else:
        word_freq[token[0]][token[1]] += 1
```

```python
df = pd.DataFrame.from_dict(word_freq).fillna(0).transpose()
df = df.div(df.sum(axis=1), axis=0) # Normalize the frequencies
df.head()
```

```python
# Select 'man' from the df
man_dist = df.loc['man']
man_dist['VB'], man_dist['NN']
```
